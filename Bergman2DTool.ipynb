## Section 10.1: Curves Defined by Parametric Equations

In this section, we learn about parametrically defined curved, i.e. curves of the form:
$$x = f(t)\quad y = g(t)$$
Our first goal is to simply familiarize ourselves with what such curves look like.
First of all, we can use parametric curves to draw the curves that come from functions.
For such curves, we will choose $x = t$ so that $y = g(t) = g(x)$ is just the graph of a function.
In the code block below, feel free to edit the function definition.
The code is written in Julia.

```julia
using Plots
```

```julia
# The values at which to evaluate the parametric function. The syntax is  smallest:step:largest.
t = -10:.1:10;
x = t;

# This is the function definition. Feel free to change this to anything you feel like! 
# You might need to first learn how to express your function correctly in Julia syntax.
y = t.^3+1;

# Finally, plot the functions

plot(x, y)
plot!(title= "A Simple Parametric Curve", ylabel= "y", xlabel="x", legend=false)
plot!(ylim=[-1, 1]*maximum(abs.(y)))
```

We can make things slightly more interesting by instead setting $y=t$ and then $x=f(t)=f(y)$, so that we have just turned our typical functions on their side.

```julia
# The values at which to evaluate the parametric function. The syntax is  smallest:step:largest.
t = -10:.1:10;
y = t;

# This is the function definition. Feel free to change this to anything you feel like! 
# You might need to first learn how to express your function correctly in Julia syntax.
x = sin.(y);

# Finally, plot the functions

plot(x, y)
plot!(title= "A Simple Parametric Curve", ylabel= "y", xlabel="x", legend=false)
plot!(xlim=[-1, 1]*maximum(abs.(x)))
```

But parametric curves are much more interesting than curves that have to pass either the vertical line test or the horizontal line test!
We can very easily create parametric curves that trace out a circle, ellipse, and much more.
Here's some examples of the interesting curves we can get with this quite powerful tool:

```julia
# Example 1
t = 0:0.01:2*pi;
r = 1;
x = r*cos.(t);
y = r*sin.(t);
plot(x,y)
plot!(title= "A Non-function Parametric Curve", xlabel = "x", ylabel = "y", legend = false)
```

```julia
# Example 2
t = 0:0.01:2*pi;
r = 1;
x = r*cos.(t);
y = r*sin.(2*t);
plot(x,y)
plot!(title= "A Non-function Parametric Curve", xlabel = "x", ylabel = "y", legend = false)
```

```julia
# Example 3
t = 0:0.01:2*pi;
x = t+sin.(5*t);
y = t+sin.(6*t);
plot(x,y)
plot!(title= "A Non-function Parametric Curve", xlabel = "x", ylabel = "y", legend = false)
```

We conclude this section with the cycloid, the favorite child of parametric curves as it actually solves certain, easy-to-state problems (see the brachistochrone problem and the tautochrone problem).
The defining function for the cycloid is
$$x = r(\theta-\sin\theta)\quad y=r(1-\cos\theta)\quad\theta\in\mathbb{R}$$
For now, we will just look at what the cycloid looks like.
The code to plot it is below:

```julia
# The Cycloid
r = 1;
n = 3;
t = 0:0.01:n*2*pi;
x = r*(t-sin.(t));
y = r*(1-cos.(t));
plot(x,y)
plot!(title= "$n Cycles of the Cycloid", xlabel = "x", ylabel = "y", legend = false)
plot!(ylim=[0,4*r])
```
